```python
import pandas as pd
import numpy as np
```

```python
data = '../data/raw_Data_Econocast.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.tail()
```

```python
df.info()
```

## Crear Cracteristica Fecha

```python
months = {
    'Enero': 'January', 'Febrero': 'February', 'Marzo': 'March', 'Abril': 'April',
    'Mayo': 'May', 'Junio': 'June', 'Julio': 'July', 'Agosto': 'August',
    'Septiembre': 'September', 'Octubre': 'October', 'Noviembre': 'November', 'Diciembre': 'December'
}

df['mes'] = df['mes'].replace(months)

df['fecha'] = pd.to_datetime(df['anio'].astype(str) + '-' + df['mes'], format='%Y-%B', errors='coerce')
```

```python
cols_numericas = ['importaciones_valor_usd', 'importaciones_volumen_kg',
                  'exportaciones_valor_usd', 'exportaciones_volumen_kg']


df[cols_numericas] = df[cols_numericas].replace({',': ''}, regex=True).astype(float)

df.tail()
```

## Eliminar las columnas vacias

```python
df = df[df['anio'] >= 2005]
```

```python
df.info()
```

```python
df.tail()
```

##  Valores Nulos

```python
df.fillna(df.median(numeric_only=True), inplace=True)

df.tail()
```

```python
df.info()
```

## Guardar Preprocesado

```python
df.to_csv('../data/preprocess_Data_EconoCast.csv', index=False)
```
