## Prueba de Estacionariedad (ADF Test)

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(df['precio_promedio_venta_USD'])
print(f"Valor p: {adf_test[1]}")  # Si p < 0.05, la serie es estacionaria
```

La serie NO es estacionaria (p = 0.96 > 0.05). Debes aplicar una diferenciación para transformarla.

```python
import time
import pickle
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

# Diccionarios de modelos
cnames = ["Random Forest", "Linear Regression", "Exponential Smoothing"]

models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Linear Regression": LinearRegression(),
    "Exponential Smoothing": ExponentialSmoothing()
}

# 4. Search Model Function
def search_model_cl(X_train, y_train, X_val, y_val, X_test, y_test, cnames=cnames, models=models):
    mse_Vali = []
    mse_Test = []
    fitting = []
    times = []

    for model_name, model in models.items():
        # ======= TRAIN ========
        start = time.time()
        model.fit(X_train, y_train)
        end = time.time()
        times.append(round(end - start, 3))

        # ===== VALIDATION =====
        y_val_pred = model.predict(X_val)
        mse_val = mean_squared_error(y_val, y_val_pred)

        # ======= TEST ========
        y_test_pred = model.predict(X_test)
        mse_test = mean_squared_error(y_test, y_test_pred)

        # ======= SCORE =======
        mse_Vali.append(round(mse_val, 3))
        mse_Test.append(round(mse_test, 3))
        fitting.append(round((mse_val - mse_test), 3))

    df_models = pd.DataFrame([mse_Vali, mse_Test, fitting, times], columns=cnames,
                             index=['MSE Vali', 'MSE Test', 'Fitting', 'Seconds'])
    df_models = df_models.sort_values(by=['MSE Vali', 'MSE Test', 'Seconds', 'Fitting'], axis=1, ascending=True)

    # Guardar el mejor modelo
    best_model = models[cnames[0]]
    with open(f'./models/best_model.pkl', 'wb') as file:
        pickle.dump(best_model, file)

    return df_models
```
